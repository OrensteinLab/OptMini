```python
def randfs(kmers,kmer,dep,w,k,sigma): #auxiliary#   recursively computes array of gamechanger probability for windows
    if dep == w:
        l = len(kmers)
        if kmer in kmers:
            probab[l-1] += 1    #there are l k-mers, probability is 1/l (only if pref is minimizer)
        else:
            probab[l] += 2  #there are (l+1) k-mers, probability is 2/(l+1) (either pref or suff is minimizer)
    else:
        nextkmer = sigma *(kmer % sigma**(k-1))
        for t in range(sigma):
            randfs(kmers|{kmer},nextkmer+t,dep+1,w,k,sigma)

def randensity(w,k,sigma): #computes the expected density of random sigma-ary order (needed for w>k)
    # calls randfs
    global probab
    probab = [0 for i in range(w+1)]    #probab[i] refers to windows with i+1 distinct k-mers
    kmers = set()
    for km in range(sigma**(k-1)):   #all k-mers starting with 0 (symmetry)
        randfs(kmers,km,0,w,k,sigma)
    #print(probab)
    probsum = probab[0]
    for ii in range(1,w+1):
        probsum += probab[ii] / (ii+1)
    return (w+1) * probsum / sigma**(w+k-1)     # density factor
```

```python
import csv

# --- Open CSV file for writing ---
with open("density_results_dfs.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["k", "w", "density_factor"])   # header

    for k in range(5, 7):           # k = 5,6
        print("k =", k)
        for w in range(2, 21):      # w = 2..20
            d = randensity(w, k, 2)
            print(" w =", w, " density factor =", d)

            # write to CSV
            writer.writerow([k, w, d])

print("Saved to density_results_dfs.csv")
```
