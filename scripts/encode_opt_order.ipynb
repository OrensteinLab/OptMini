```python
string_min = "55 64 6 5 44 2 32 4 16 45 64 27 18 31 36 3 64 46 43 47 12 54 29 1 8 33 13 37 21 38 39 0 64 23 48 30 49 15 50 7 10 51 64 11 9 28 40 24 64 52 17 34 53 64 14 41 64 20 19 35 22 42 25 26"
```

```python
# make it into an array
opt_order = [int(x) for x in string_min.split(" ")]

# print to verify
print(opt_order)
```

```python
# all 64 vlaues should be given the minimum + 1 of the etnrire array
n_kmers = 32 
new_arr = []
counter = 0
for i in opt_order:
    if i == n_kmers:
        new_arr.append(n_kmers-counter)
        counter+=1
    else:
        new_arr.append(i)

print(new_arr)
```

```python
import struct

def save_vector_to_file(filename, vector):
    with open(filename, 'wb') as file:
        # Write size as size_t (platform-dependent)
        file.write(struct.pack('P', len(vector)))
        
        # Write all values as uint64_t
        file.write(struct.pack(f'{len(vector)}Q', *vector))


# encode it
save_vector_to_file('opt_order.gm', new_arr)
```

```python



def load_vector_from_file(filename):
    with open(filename, 'rb') as file:
        # Read the size of the vector (stored as size_t in C++, which is platform-dependent)
        size_data = file.read(struct.calcsize('P'))  
        size = struct.unpack('P', size_data)[0]
        
        # Read the rest of the data as uint64_t values
        vector_data = file.read(size * struct.calcsize('Q')) 
        vector = struct.unpack(f'{size}Q', vector_data)
        
        return list(vector)


filename = 'opt_order.gm'
vector = load_vector_from_file(filename)
print(vector)
print(len(vector))
```
